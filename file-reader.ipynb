```python
import pandas as pd

column_names = [
    "ent_num",
    "Company Name",
    "SDN_Type",
    "Program",
    "Title",
    "Call_Sign",
    "Vess_type",
    "Tonnage",
    "GRT",
    "Vess_flag",
    "Vess_owner",
    "Remarks"
]

prim_df = pd.read_csv("cons_prim.csv")
prim_df.columns = column_names
print(prim_df.head())
```

```python
prim_df = prim_df[prim_df["SDN_Type"]).str.lower() != "individual"]
prim_df = prim_df[prim_df["SDN_Type"].str.lower() != "individual"]
```

```python
add_df = pd.read_csv("cons_add.csv")
add_df.columns = ['ent_num', 'Add_num', 'Address', 'City_State_postal_code', 'Country', 'Add_remarks']
add_df.head()
```

```python
prim_df.head()
```

```python
merge_df = pd.merge(prim_df[['ent_num', 'Company Name', 'Program']], 
                    add_df[['ent_num', 'Country']].drop_duplicates(), 
                    on='ent_num')
china_companies = merge_df[(merge_df['Country'].str.lower().str.contains ('china|hong kong')) & 
                           (merge_df['Program'].str.lower().str.contains('cmic'))]
print( 'shape: ', china_companies.shape)
china_companies.head()
```

```python
china_companies.to_csv('china_companies.csv', index=False)
```
